```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
import joblib
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix
```

```python
df = pd.read_csv('/content/drive/MyDrive/titanic.csv', sep=',')
```

```python
def prepare_data(df, test_size=0.2):
    # select features
    feature = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
    target = 'Survived'

    # separate features and target
    train_features = df[feature].copy()
    train_target = df[target].copy()

    # split data to train and test
    X_train, X_test, y_train, y_test = train_test_split(train_features, train_target, test_size=test_size, random_state=42, stratify=train_target)

    #### Training data ####
    # fill missing value
    age_imputer = SimpleImputer(strategy='median')
    new_age = age_imputer.fit_transform(X_train[['Age']])
    X_train['Age'] = new_age.flatten()

    # export imputer
    export_object(age_imputer, output_name="age_imputer.joblib")

    # encode sex
    sex_labenc = LabelEncoder()
    encoded_sex = sex_labenc.fit_transform(X_train['Sex'])
    X_train['Sex'] = encoded_sex

    # export sex encoder
    export_object(sex_labenc, output_name="sex_labenc.joblib")

    # encode embarked
    embarked_labenc = LabelEncoder()
    encoded_embarked = embarked_labenc.fit_transform(X_train['Embarked'])
    X_train['Embarked'] = encoded_embarked

    # export embarked encoder
    export_object(embarked_labenc, output_name="embarked_labenc.joblib")


    #### Testing data ####
    test_new_age = age_imputer.transform(X_test[['Age']])
    test_new_age
    X_test['Age'] = test_new_age.flatten()

    test_encoded_sex = sex_labenc.transform(X_test['Sex'])
    X_test['Sex'] = test_encoded_sex

    test_encoded_embarked = embarked_labenc.transform(X_test['Embarked'])
    X_test['Embarked'] = test_encoded_embarked

    return X_train, X_test, y_train, y_test
```

```python
def evaluate(labels, predictions, classes=None):
    print(f"Accuracy: {accuracy_score(labels, predictions):.3f}")
    print(f"F1 Score: {f1_score(labels, predictions, average='micro'):.3f}")
    print(classification_report(labels, predictions))

    cm = confusion_matrix(labels, predictions, labels=classes)
    disp = ConfusionMatrixDisplay(cm, display_labels=classes)
    disp.plot()
    plt.show()
```

```python
def train_evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)

    predictions = model.predict(X_test)

    evaluate(y_test, predictions, classes=model.classes_)

    return model
```

```python
def export_object(obj, output_name):
    joblib.dump(obj, output_name)
```

```python
def run_ml(model, data_path, test_size=0.2, output_name="model.joblib"):
    # read data
    df = read_data(data_path)

    # preprocess data
    X_train, X_test, y_train, y_test = prepare_data(df, test_size)

    # train model
    model = train_evaluate_model(model, X_train, X_test, y_train, y_test)

    # export model
    export_object(model, output_name=output_name)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)

print('KNN Model')
print()
run_ml(knn, data_path='/content/drive/MyDrive/titanic.csv', test_size=0.2, output_name="knn.joblib")
```

```python
dtree = DecisionTreeClassifier(random_state=42)

print('Decision Tree Model')
print()
run_ml(dtree, data_path='/content/drive/MyDrive/titanic.csv', test_size=0.2, output_name="dtree.joblib")
```

```python
svc = SVC(random_state=42)

print('SVM Model')
print()
run_ml(svc, data_path='/content/drive/MyDrive/titanic.csv', test_size=0.2, output_name="svc.joblib")
```

```python
rfc = RandomForestClassifier(n_estimators=100, random_state=0)

print('Random Forest Model')
print()
run_ml(rfc, data_path='/content/drive/MyDrive/titanic.csv', test_size=0.2, output_name="rfc.joblib")
```

```python
xgb = XGBClassifier(n_estimators=100, random_state=56)

print('XGB Classifier')
print()
run_ml(xgb, data_path='/content/drive/MyDrive/titanic.csv', test_size=0.2, output_name="xgb.joblib")
```
